# 2. Weekly graph question

Suppose you plot the number visitors to a museum over the seven days of a week; then again the following week.  Can you suggest the advantages and disadvantages of the following two ways of presenting the data?

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
time_series_0 = np.array([100, 110, 130, 120, 100, 90, 80])
time_series_1 = np.array([70, 90, 100, 100, 120, 110, 90])
fig, axs = plt.subplots(1, 2)
axs[0].plot(time_series_0)
axs[1].plot(time_series_1)
plt.subplots_adjust(wspace = 0)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

```python
plt.plot(time_series_0)
plt.plot(time_series_1)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

When presenting the data using line graphs, an advantage is that it allows viewers to distinctly analyze each week's trends without visual clutter. This format simplifies the process of identifying day to day fluctuations and makes it easy to assess individual week patterns in isolation. However, this approach has limitations that include the challenge of interpreting broader trends or relationships between the weeks. Separating the data can hinder the ability to visualize continuity or observe inter-week interactions effectively, thus requiring viewers to mentally bridge the two datasets.

Conversely, overlaying both time series on a single graph provides a consolidated view, making comparisons between the two weeks more straightforward. This format highlights intersections, peak alignments, and disparities within one single, cohesive visualization. It also minimizes visual separation and facilitates trend interpretation. Nonetheless, this method introduces potential drawbacks, such as overlapping data points, that can obscure individual trends. Also, if the visualization lacks clear differentiation, such as distinct colors or markers, viewers may struggle to follow each dataset independently, therefore diminishing clarity for week-specific insights.

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 187-205 of the Storytelling With Data book as best you can. (Chapter Eight). You do not have to get the exact data values right, just the overall look and feel.

```python
import matplotlib.pyplot as plt

years = [2008, 2009, 2010, 2011, 2012, 2013, 2014]
prices_A = [400, 420, 430, 400, 350, 320, 300]
prices_B = [380, 400, 410, 380, 340, 310, 300]

plt.figure(figsize=(8, 5))

plt.plot(years, prices_A, label='Item A', color='#1f77b4', linestyle='-', linewidth=2)
plt.plot(years, prices_B, label='Item B', color='#1f77b4', linestyle='-', linewidth=2)

plt.title('Retail price over time', fontsize=16, weight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average price ($)', fontsize=12)

plt.xticks(years, fontsize=10)
plt.yticks(range(0, 501, 100), fontsize=10)
plt.xlim(2008, 2014)
plt.ylim(0, 500)

plt.grid(visible=True, linestyle='--', alpha=0.5)

plt.text(2009, 425, 'A', color='#1f77b4', fontsize=10, fontweight='bold')
plt.text(2009, 405, 'B', color='#1f77b4', fontsize=10, fontweight='bold')

plt.legend(loc='upper right', fontsize=10)

plt.figtext(0.1, 0.02, 'FIGURE 8.13', wrap=True, fontsize=10, weight='bold')

plt.tight_layout()
plt.show()
```
