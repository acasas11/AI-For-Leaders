# 2. Weekly graph question

Below are the graphs from last week's weekly graph question.  Suppose you wanted to cause the viewer to examine the last three months of data first.  How could you use preattentive attributes to do that?  Suggest several possibilities.

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

Possibilities:
Using preattentive attributes to focus attention on the last three months of data can be a good way to guide the viewers analysis. Color is a powerful tool for this. By highlighting the most recent data points or segments in colors that are nright and contrasting, like red or orange, while keeping the rest of the data in more muted tones, like grey or blue, the viewers eye can be immediately drawned to the area of interest. This would create a visual hierarchy where the last data from the last three months stands out prominently against the backdrop of the earlier data.
Size is another attribute that can be utilized to highlight specific data points. Increasing the size of markers and bars that represent the last three months of data would make them more prominent. For line graphs, thickening the line for the last three months can also create a more clear visual distinction. This use of size variation would help the viewer to identify this newer data more quickly and focus on the highlighted period without needing to consciously search for it.
Also, position and shape can further enhance the focus on the desired data. By isolating the last three months of data by creating a shaded background or a distinct boundary around them, the new data becomes separated from the rest and thus makes it easier for the viewer to notice. Using unique shapes, like triangles for the most recent data points compared to circles for the rest, adds another layer of differentiation.
These preattentive attributes work together to create a visual emphasis on the last three months worth of data, effectively guiding the viewers attention.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Country': ['AUS', 'BRA', 'CAN', 'CHI', 'FRA', 'GER', 'IND', 'ITA', 'MEX', 'RUS', 'SPA', 'TUR', 'UK', 'US'],
    'A': [1, 1, 2, 1, 3, 3, 4, 2, 1, 4, 2, 7, 1, 1],
    'B': [2, 3, 3, 2, 2, 1, 1, 4, 5, 3, 3, 2, 2, 2],
    'C': [3, 4, 6, 8, 4, 6, 8, 10, 4, 7, 4, 3, 3, 4],
    'D': [6, 5, 12, 4, 8, 5, 10, 9, 6, 9, 5, 4, 6, 3],
    'E': [7, 6, 8, 7, 10, 4, 5, 8, 3, 12, 11, 8, 7, 5],
}

df = pd.DataFrame(data)
df.set_index('Country', inplace=True)

palette = sns.color_palette("Spectral", as_cmap=True)

plt.figure(figsize=(10, 6))
ax = sns.heatmap(df, cmap=palette, annot=True, cbar=False, linewidths=0.5, linecolor='black', fmt="d")
ax.set_title("Sales Rank by Country and Product")

plt.show()
```
